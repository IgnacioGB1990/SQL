```python
# Libraries
import pandas as pd
import sqlite3

cnx = sqlite3.connect('./data/jobs.db')

# Definimos la función para hacer queries.
def sql_query(query):
    return pd.read_sql(query, cnx)
```

```python
query = """
SELECT *
FROM jobs


"""
sql_query(query)
```

### Without RegExp

```python
query = """

SELECT *
FROM jobs
WHERE
    lower(Name) LIKE "%JOHN%" OR
    lower(Name) LIKE "%Carlos%" OR
    lower(Name) LIKE "%Anderson%" OR
    lower(Name) LIKE "%Snow%" 

"""
sql_query(query)
```

```python
query = """

SELECT *
FROM jobs
WHERE
    REGEXP_LIKE(lower(Name),'JOHN|Carlos|Anderson|Snow')

"""
sql_query(query)
```
