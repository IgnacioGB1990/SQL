```python
# Libraries
import pandas as pd
import sqlite3

cnx = sqlite3.connect('./data/jobs.db')

# Definimos la función para hacer queries.
def sql_query(query):
    return pd.read_sql(query, cnx)
```

```python
query = """
SELECT *
FROM jobs


"""
sql_query(query)
```

### Without RegExp

```python
query = """

SELECT *
FROM jobs
WHERE
    lower(Name) LIKE "%JOHN%" OR
    lower(Name) LIKE "%Carlos%" OR
    lower(Name) LIKE "%Anderson%" OR
    lower(Name) LIKE "%Snow%" 

"""
sql_query(query)
```

### `Subquery`



```python
SELECT title, rental_rate
FROM film
WHERE rental_rate >
(SELECT AVG(rental_rate) FROM film)
```

```python
SELECT film_id, title
FROM film
WHERE film_id IN

(SELECT inventory.film_id
FROM rental
INNER JOIN inventory ON inventory.inventory_id = rental.inventory_id
WHERE return_date BETWEEN '2005-05-29'AND '2005-05-30')
ORDER BY film_id ASC
```

```python
SELECT first_name, last_name
FROM customer AS c
WHERE EXISTS
(SELECT * FROM payment AS p
WHERE p.customer_id = c.customer_id
AND amount >11)
```

### `Self-join`

- It is a query in which a table is joined to itself.
- Are useful for comparing values in a column of rows within the same table.
- The self join can be viewed as a joint of two copies of the same table.
- There is no special keyword for a self join, its simply standard JOIN syntax with the same table in both parts.
